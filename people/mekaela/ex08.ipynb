```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
df = pd.read_csv('./EFIplus_medit.zip',compression='zip', sep=";")
df.head(5)
```

From last exercise:

```python
species_columns = df.columns[df.columns.get_loc('Abramis brama'):]

# sum each row
df['species_richness'] = df[species_columns].sum(axis=1)

print(df.head())
```

```python
# List of continuous environmental variables
continuous_vars = ['Altitude', 'Actual_river_slope', 'Elevation_mean_catch', 'prec_ann_catch', 'temp_ann', 'temp_jan', 'temp_jul']

# Remove rows with NA values
df = df.dropna(subset=continuous_vars)
```

```python
# for the later question, a set of regression coeffients for each var
simple_regression_coefficients = {}

for var in continuous_vars:
    #replace zeros with small positive value to enable log transformation
    if df[var].min() <=0:
        df[var] = np.where(df[var]<=0, 0.000001, df[var])

    # Check skewness and apply log transformation if necessary
    if df[var].skew() > 1 :
        df[var + "_log"] = np.log(df[var])
        var_to_use = var + "_log"
    else:
        var_to_use = var

    # Simple Linear Regression
    X = df[var_to_use]
    y = df['species_richness']  # Assuming 'Species richness' is the target variable
    X = sm.add_constant(X) #adds a constant term to the predictor
    model = sm.OLS(y, X).fit()
    simple_regression_coefficients[var] = model.params[1]  # Extract the coefficient for the predictor

    # Print results
    print(f"Regression results for {var}:")
    print(model.summary())

    # Plot fitted model
    plt.figure(figsize=(8, 6))
    plt.scatter(X[var_to_use], y, label='Observed data')
    plt.plot(X[var_to_use], model.fittedvalues, color='red', label='Fitted model')
    plt.xlabel(var_to_use)
    plt.ylabel('Species richness')
    plt.title(f'Simple Linear Regression: Species richness vs {var}')
    plt.legend()
    plt.show()
```

```python
# Multiple Linear Regression
X = df[continuous_vars]
y = df['species_richness']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()

print(model.summary())
```

```python
# Compare coefficients
print("\nComparison of Coefficients:")

# Access coefficients from the multiple linear regression model
multiple_regression_coefficients = model.params[1:]  # Exclude the intercept

# Create a DataFrame for comparison
comparison_df = pd.DataFrame({
    'Variable': continuous_vars,
    'Simple Regression Coefficient': [simple_regression_coefficients.get(var, np.nan) for var in continuous_vars],
    'Multiple Regression Coefficient': multiple_regression_coefficients.values
})

print(comparison_df)
```

Plot the partial dependence plots to compare responses with the previous single effects.

```python
fig = sm.graphics.plot_partregress_grid(model)
fig.tight_layout(pad=1.0)
```

## Check the multicolinearity among the predictor variables and run a second more parsimonious regression model. Check how the coefficients have changed.

```python
# Calculate VIF (Variance Inflation Factor)
X = df[continuous_vars]
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]

print(vif_data)

# Remove variables with high VIF (e.g., VIF > 5 or 10)
# Example: Removing 'Elevation_mean_catch' and 'prec_ann_catch' due to high VIF
variables_to_keep = ['Altitude', 'Actual_river_slope', 'temp_ann', 'temp_jan', 'temp_jul']
X2 = df[variables_to_keep]

# Fit the second regression model
X2 = sm.add_constant(X2)
model2 = sm.OLS(y, X2).fit()
print(model2.summary())

# Compare coefficients
print("\nOriginal Model Coefficients:")
print(model.params)

print("\nSecond Model Coefficients:")
print(model2.params)
```

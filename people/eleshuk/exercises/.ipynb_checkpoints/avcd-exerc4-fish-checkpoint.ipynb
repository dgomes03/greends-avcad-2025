Test, using both visualization and hypothesis testing methods, if the actual_river_slope is drawn from a
normal distribution.

```python
# Actual_river_slope
stats.probplot(df['Actual_river_slope'], dist="norm", plot=plt)
plt.title('Q-Q Plot of Actual_river_slope')
plt.show()
```

```python
stat, p_value = shapiro(df['Actual_river_slope'])
print(f"Shapiro-Wilk Test: Statistics={stat:.3f}, p-value={p_value:.3f}")
if p_value > 0.05:
    print('Fail to reject H0: Data seems normally distributed')
else:
    print('Reject H0: Data does not seem normally distributed')
```

```python
# take 100 random samples of 2000 observations each of Actual_River_Slope (with replacement), compute the mean of each sample and check 
# the distribution of these 100 mean values

sample_means = []

# 100 samples of size 2000, with replacement
for i in range(100):
    sample = df['Actual_river_slope'].sample(n=2000, replace=True)
    sample_mean = sample.mean()
    sample_means.append(sample_mean)

# Convert to numpy array or pandas Series if needed
sample_means = np.array(sample_means)

# Check distribution
plt.hist(sample_means, bins=20, edgecolor='k')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.title('Distribution of Sample Means')
plt.show()
```

```python
stat, p_value = shapiro(sample_means)
print(f"Shapiro-Wilk Test p-value: {p_value}")
```

p-value < 0.05 - reject normality? 
Q-Q and histogram visually look normal?
